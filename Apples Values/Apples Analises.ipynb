Foi identificado que existe valores não numéricos no array, para identificar quantos itens existem podemos usar a seguinte forma?

```python
sum(np.isnan(Kaliningrad))
```

Para corrigir esse valor, podemos interpolar esse valor (valor médio entre a soma do valor do antecessor e do sucessor)

```python
# Kaliningrad[2] = 'nan'  # Forçando a posição 3 a ser NaN. Remova o primeiro '#' caso deseje rodar esta linha.
nanLocations = np.where(np.isnan(Kaliningrad))[0]

# Verifica se há um ou mais NaNs
if len(nanLocations) > 0: 
    # Cria um laço nas posições encontradas, substituindo o valor NaN pela média dos valores anterior e posterior.
    for nanLocation in nanLocations:
        Kaliningrad[nanLocation] = (Kaliningrad[nanLocation - 1] + Kaliningrad[nanLocation + 1]) / 2

# Mostra os valores atualizados
Kaliningrad
```

Calculando a Regressão Lenear

O cálculo base da reta é
y = ax+b

O calculo da regressão lenear é feito da seguinte maneira:

Onde:
â = coeficiente angular;
n = número de elementos;
Y = Moscow;
X = datas.

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

Aplicando a formula

```python
a = ((n*np.sum(X*Y)) - (np.sum(X) * np.sum(Y))) / (n*np.sum(X**2) - (np.sum(X))**2)
a
```

Calculando o Coeficiente Linear (b)

```python
b = np.mean(Y) - a*np.mean(X)
```

Temos então os valores de â e b, com isso é possivel calcular o y

```python
y = a*X+b
```

Plotando no gráfico

```python
plt.plot(datas,Moscow)
plt.plot(datas,y)
```
