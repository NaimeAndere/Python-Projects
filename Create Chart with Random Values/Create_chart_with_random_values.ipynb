Gráfico de notas de matemática

```python
#Import módulo Random
from random import randrange

#Import módulo MatplotLib
import matplotlib.pyplot as plt
```

Adicionando Valores para X e Y

```python
#Adicionar valores para X e Y
x = list(range(1,9))

notas_matematica = []

for notas in range(8):
    notas_matematica.append([randrange(0,11)])
y = notas_matematica
```

Montando o Gráfico

```python
# Identifica os valores de X e Y, e adiciona 'o' como marcador
plt.plot(x, y, marker='o')

# Adiciona título para o gráfico
plt.title('Notas de matemática')
#Adiciona nomes aos eixos X e Y
plt.xlabel('Provas')
plt.ylabel('Notas')

#chama a visualização do gráfico
plt.show()
```

Criando uma lista com 3 máterias e suas notas, e montando 1 gráfico para cada, usando a função 'for'

```python
notas_matematica = ['Matemática',8,7,6,6,7,7,8,10]
notas_portugues = ['Português',9,9,9,8,5,6,8,5]
notas_geografia = ['Geografia',10,10,6,7,7,7,8,7]

notas = [notas_matematica, notas_portugues, notas_geografia]
```

```python
for nota in notas:
    x = list(range(1, 9))
    y = nota[1:]
    plt.plot(x, y, marker='o')
    plt.xlabel('Provas')
    plt.ylabel('Notas')
    plt.title(nota[0])
    plt.show()
```
